## <span style="color: yellow;">**Pengantar AI**</span>

AI memungkinkan kita membangun perangkat lunak luar biasa yang dapat meningkatkan perawatan kesehatan, memungkinkan orang mengatasi kekurangan fisik, memberdayakan infrastruktur cerdas, menciptakan pengalaman hiburan yang luar biasa, dan bahkan menyelamatkan planet ini!

Apa itu AI?
Sederhananya, AI adalah perangkat lunak yang meniru perilaku dan kemampuan manusia. Beban kerja utama meliputi:

- Pembelajaran mesin - Ini sering menjadi fondasi untuk sistem AI, dan merupakan cara kita "mengajari" model komputer untuk membuat prediksi dan menarik kesimpulan dari data.
- Visi komputer - Kemampuan dalam AI untuk menafsirkan dunia secara visual melalui kamera, video, dan gambar.
- Pemrosesan bahasa alami - Kemampuan dalam AI untuk komputer menginterpretasikan bahasa tertulis atau lisan, dan merespons dalam jenis.
- Kecerdasan dokumen - Kemampuan dalam AI yang menangani pengelolaan, pemrosesan, dan penggunaan data dalam volume tinggi yang ditemukan dalam formulir dan dokumen.
- Penambangan pengetahuan - Kemampuan dalam AI untuk mengekstrak informasi dari volume besar data yang sering tidak terstruktur untuk membuat penyimpanan pengetahuan yang dapat dicari.
- AI Generatif - Kemampuan dalam AI yang membuat konten asli dalam berbagai format termasuk bahasa alami, gambar, kode, dan banyak lagi.


## <span style="color: yellow;">**Memahami Pembelajaran Mesin**</span>

Pembelajaran Mesin adalah fondasi untuk sebagian besar solusi AI. Sejak tahun 1950-an, para peneliti, sering dikenal sebagai ilmuwan data, telah bekerja pada pendekatan yang berbeda untuk AI. Sebagian besar aplikasi modern AI memiliki asal-usul mereka dalam pembelajaran mesin, cabang AI yang menggabungkan ilmu komputer dan matematika.

Mari kita mulai dengan melihat contoh dunia nyata tentang bagaimana pembelajaran mesin dapat digunakan untuk memecahkan masalah yang sulit.

Teknik pertanian berkelanjutan sangat penting untuk memaksimalkan produksi pangan sekaligus melindungi lingkungan yang rapuh. Yield, perusahaan teknologi pertanian yang berbasis di Australia, menggunakan sensor, data, dan pembelajaran mesin untuk membantu petani membuat keputusan berdasarkan informasi terkait kondisi cuaca, tanah, dan tanaman.

### Cara kerja pembelajaran mesin
Jadi, bagaimana mesin belajar?

Jawabannya adalah, dari data. Di dunia saat ini, kita membuat data dalam jumlah besar saat kita menjalani kehidupan sehari-hari. Mulai dari pesan teks, surel, dan kiriman media sosial yang kita kirim untuk foto dan video yang kita ambil di ponsel, kita menghasilkan sejumlah besar informasi. Lebih banyak data masih dibuat oleh jutaan sensor di rumah kita, mobil, kota, infrastruktur transportasi umum, dan pabrik.

Ilmuwan data dapat menggunakan seluruh data tersebut untuk melatih model pembelajaran mesin yang dapat membuat prediksi dan kesimpulan berdasarkan hubungan yang mereka temukan dalam data.

Model pembelajaran mesin mencoba menangkap hubungan antara data. Misalnya, organisasi konservasi lingkungan ingin sukarelawan mengidentifikasi dan membuat katalog spesies bunga liar yang berbeda menggunakan aplikasi ponsel. Animasi berikut menunjukkan bagaimana pembelajaran mesin dapat digunakan untuk mengaktifkan skenario ini.

- Tim ahli botani dan ilmuwan mengumpulkan data tentang sampel bunga liar.
- Tim melabeli sampel dengan spesies yang benar.
- Data berlabel diproses menggunakan algoritma yang menemukan hubungan antara fitur sampel dan spesies berlabel.
- Hasil algoritma dirangkum dalam model.
- Ketika sampel baru ditemukan oleh sukarelawan, model dapat mengidentifikasi label spesies yang benar.
Pendekatan ke AI memiliki tingkat lanjut untuk menyelesaikan tugas dengan kompleksitas yang jauh lebih besar. Model kompleks ini membentuk dasar kemampuan AI.

## <span style="color: yellow; font-weight: bold;">Tantangan dan resiko dengan AI</span>
Kecerdasan Buatan adalah alat canggih yang dapat digunakan untuk memberikan manfaat terbaik bagi dunia. Namun, seperti alat apa pun, alat harus digunakan secara bertanggung jawab.

### Solusi untuk Mengatasi Tantangan Ini:
Untuk mengatasi tantangan dan risiko tersebut, penting bagi pengembang AI untuk:

- Memperbaiki bias dalam data: Melakukan audit dan pembersihan data secara rutin, serta memastikan data yang digunakan untuk melatih model AI representatif dan bebas bias.
- Mengutamakan keamanan: Melindungi data sensitif dengan enkripsi dan sistem keamanan yang kuat untuk mencegah kebocoran data.
- Menjamin aksesibilitas: Merancang solusi yang dapat diakses oleh semua orang, termasuk mereka yang memiliki kebutuhan khusus.
- Meningkatkan transparansi: Membuka proses pembuatan keputusan AI dan menjelaskan dasar dari keputusan tersebut, sehingga pengguna dapat lebih memahami bagaimana teknologi bekerja.
- Menetapkan tanggung jawab yang jelas: Menyusun pedoman hukum yang jelas mengenai siapa yang bertanggung jawab ketika AI membuat kesalahan atau menyebabkan kerugian.

## <span style="color: yellow; font-weight: bold;">Memahami AI yang bertanggung jawab</span>
Di Microsoft, pengembangan perangkat lunak AI dipandu oleh serangkaian enam prinsip, yang dirancang untuk memastikan aplikasi AI memberikan solusi luar biasa untuk masalah yang sulit tanpa konsekuensi negatif yang tidak diinginkan.

### **Keadilan**
Sistem AI harus memperlakukan semua orang secara adil. Misalnya, anggap Anda membuat model pembelajaran mesin untuk mendukung aplikasi persetujuan pinjaman untuk bank. Model harus memprediksi apakah pinjaman harus disetujui atau ditolak tanpa bias. Bias ini dapat didasarkan pada jenis kelamin, etnis, atau faktor lain yang menghasilkan keuntungan atau kerugian yang tidak adil bagi kelompok pelamar tertentu.

Azure Machine Learning mencakup kemampuan untuk menafsirkan model dan mengukur sejauh mana setiap fitur data memengaruhi prediksi model. Kemampuan ini membantu ilmuwan dan pengembang data mengidentifikasi dan mengurangi bias dalam model.

Contoh lain adalah implementasi Microsoft dari [<span style="color: light-blue;">AI yang Bertanggung Jawab dengan Layanan Wajah</span>](https://azure.microsoft.com/en-us/blog/responsible-ai-investments-and-safeguards-for-facial-recognition/), yang menghentikan kemampuan pengenalan wajah yang dapat digunakan untuk mencoba menyimpulkan status emosional dan atribut identitas. Kemampuan ini, jika disalahgunakan, dapat membuat orang tunduk pada stereotip, diskriminasi, atau penolakan layanan yang tidak adili.

### **Keandalan dan Kemanan**
Sistem AI harus berperforma andal dan aman. Misalnya, pertimbangkan sistem perangkat lunak berbasis AI untuk kendaraan otonom; atau model pembelajaran mesin yang mendiagnosis gejala pasien dan merekomendasikan resep. Ketidakandalan dalam sistem semacam ini dapat mengakibatkan risiko besar bagi kehidupan manusia.

Pengembangan aplikasi perangkat lunak berbasis AI harus tunduk pada pengujian yang ketat dan proses manajemen penyebaran untuk memastikan aplikasi berfungsi seperti yang diharapkan sebelum rilis.

### **Privasi dan Keamanan**
Sistem AI harus aman dan menghormati privasi. Model pembelajaran mesin yang mendasarkan sistem AI pada data bervolume besar, yang mungkin berisi detail pribadi yang harus dijaga kerahasiaannya. Bahkan setelah model dilatih dan sistem dalam produksi, privasi dan keamanan perlu dipertimbangkan. Karena sistem menggunakan data baru untuk membuat prediksi atau mengambil tindakan, data dan keputusan yang dibuat dari data mungkin tunduk pada masalah privasi atau keamanan.

### **Inklusivitas** 
Sistem AI harus memberdayakan semua orang dan melibatkan orang. AI harus membawa manfaat bagi seluruh lapisan masyarakat, terlepas dari kemampuan fisik, jenis kelamin, orientasi seksual, etnis, atau faktor lainnya.

### **Transparansi**
Sistem AI harus dapat dipahami. Pengguna harus menyadari sepenuhnya tujuan sistem, cara kerjanya, dan batasan apa yang mungkin diharapkan.

### **Akuntabilitas**
Orang harus bertanggung jawab atas sistem AI. Perancang dan pengembang solusi berbasis AI harus bekerja dalam kerangka kerja pemerintahan dan prinsip organisasi yang memastikan solusi memenuhi standar etika dan hukum yang jelas.

Prinsip-prinsip AI yang bertanggung jawab dapat membantu Anda memahami beberapa tantangan yang dihadapi pengembang saat mereka mencoba menciptakan solusi AI yang pantas.
